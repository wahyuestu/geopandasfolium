```python
import geopandas as gpd #Untuk memasukkan modul geopandas dengan menggunakan alias gpd. 
```

```python
gdf = gpd.read_file("RBI25K_ADMINISTRASI_AR.geojson") #digunakan untuk membaca data geospasial dari file dengan format GeoJSON/Json
gdf.head() #Digunakan untuk menampilkan beberapa baris pertama dari GeoDataFrame. Fungsi head() umumnya digunakan untuk memberikan gambaran cepat tentang struktur dan isi dari GeoDataFrame atau DataFrame
# gdf.info()
```

```python
gdf = gdf[['NAMOBJ', 'FCODE', 'SHAPE_AREA', 'SHAPE_LEN', 'geometry']].fillna(0) #digunakan untuk membuat subset dari GeoDataFrame yang hanya memuat kolom-kolom yang ingin dimasukkan
```

```python
pip install folium matplotlib mapclassify #untuk menginstall library folium (pembuatan peta interaktif menggunakan Leaflet.js.), matplotlib (library untuk membuat visualisasi statis dan dinamis di Python), mapclassify (library yang menyediakan alat untuk klasifikasi data spasial)
```

```python
import folium #memasukkan library folium
m = gdf.explore(column="SHAPE_AREA", name="Shape Area") #m adalah variabel, gdf (untuk memberikan warna pada elemen peta), explore (untuk menampilkan visual peta dari data geospasial)
folium.LayerControl().add_to(m) #memungkinkan pengguna untuk mengonfigurasi tampilan peta dengan memilih lapisan yang ingin ditampilkan 
m
```
